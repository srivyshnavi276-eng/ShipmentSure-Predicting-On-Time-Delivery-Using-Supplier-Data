```python
plt.figure(figsize=(6,4))
sns.countplot(x="Reached.on.Time_Y.N", data=df)
plt.title("On-Time vs Delayed Shipments")
plt.show()
```

```python
num_cols = df.select_dtypes(include='number').columns

for col in num_cols:
    print(f"\n Statistics for {col}")
    print("Mean:", df[col].mean())
    print("Median:", df[col].median())
    print("Mode:", df[col].mode()[0])
    print("Std Dev:", df[col].std())
    print("Variance:", df[col].var())
```

```python
group_stats = df.groupby("Reached.on.Time_Y.N").agg(
    Mean_Discount=("Discount_offered", "mean"),
    Median_Discount=("Discount_offered", "median"),
    Std_Discount=("Discount_offered", "std"),
    Var_Discount=("Discount_offered", "var")
)

print(group_stats)
```

```python
def detailed_stats(column):
    return {
        "Mean": df[column].mean(),
        "Median": df[column].median(),
        "Mode": df[column].mode()[0],
        "Std Dev": df[column].std(),
        "Variance": df[column].var(),
        "Min": df[column].min(),
        "Max": df[column].max(),
        "Range": df[column].max() - df[column].min()
    }

print(detailed_stats("Weight_in_gms"))
```

```python
stats_table = pd.DataFrame(columns=["Mean","Median","Mode","Std Dev","Variance"])

for col in num_cols:
    stats_table.loc[col] = [
        df[col].mean(),
        df[col].median(),
        df[col].mode()[0],
        df[col].std(),
        df[col].var()
    ]

print(stats_table)
```

```python
plt.figure(figsize=(8,5))

means = df[num_cols].mean()
medians = df[num_cols].median()

plt.plot(means.values, label="Mean", marker='o')
plt.plot(medians.values, label="Median", marker='s')

plt.xticks(range(len(num_cols)), num_cols, rotation=90)
plt.legend()
plt.title("Mean vs Median Comparison")
plt.show()
```

```python
for col in num_cols:
    print(f"{col} Skewness:", df[col].skew())
    print(f"{col} Kurtosis:", df[col].kurt(), "\n")
```

```python
cv_table = {}

for col in df.select_dtypes(include='number').columns:
    mean = df[col].mean()
    std = df[col].std()
    cv = (std / mean) * 100
    cv_table[col] = cv

print("Coefficient of Variation (%):")
print(pd.Series(cv_table))
```

```python
iqr_table = {}

for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    iqr_table[col] = IQR

print("Interquartile Range:")
print(pd.Series(iqr_table))
```

```python
percentiles = df.describe(percentiles=[0.25,0.5,0.75,0.90])
print(percentiles)
```

```python
outlier_counts = {}

for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_counts[col] = len(outliers)

print("Outlier Count per Column:")
print(pd.Series(outlier_counts))
```

```python
for col in df.select_dtypes(include='number').columns:
    skew = df[col].skew()
    
    if skew > 0:
        print(col, " Positively Skewed")
    elif skew < 0:
        print(col, " Negatively Skewed")
    else:
        print(col, " Symmetrical")
```

```python
cov_matrix = df.cov(numeric_only=True)
print("Covariance Matrix:")
print(cov_matrix)
```

```python
target_stats = df.groupby("Reached.on.Time_Y.N").describe()
print(target_stats)
```

```python
range_table = {}

for col in df.select_dtypes(include='number').columns:
    range_table[col] = df[col].max() - df[col].min()

print("Range of Columns:")
print(pd.Series(range_table))
```

```python
mad_table = {}

for col in df.select_dtypes(include='number').columns:
    mad_table[col] = np.median(np.abs(df[col] - np.median(df[col])))

print("Median Absolute Deviation:")
print(pd.Series(mad_table))
```

```python
advanced_stats = pd.DataFrame(columns=[
    "Mean","Median","Mode","Std","Variance","Skewness","Kurtosis","Range","IQR"
])

for col in df.select_dtypes(include='number').columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    
    advanced_stats.loc[col] = [
        df[col].mean(),
        df[col].median(),
        df[col].mode()[0],
        df[col].std(),
        df[col].var(),
        df[col].skew(),
        df[col].kurt(),
        df[col].max()-df[col].min(),
        Q3 - Q1
    ]

print(advanced_stats)
```
