```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

df = pd.read_csv("Train.csv")

# Define numeric columns (excluding target if needed)
num_cols = df.select_dtypes(include='number').columns
num_cols = num_cols.drop("Reached.on.Time_Y.N")  # optional (removes target column)

print(num_cols)
```

```python
for col in num_cols:
    plt.figure(figsize=(7,4))
    sns.boxplot(x="Reached.on.Time_Y.N", y=col, data=df)
    plt.title(f"{col} vs On-Time Delivery")
    plt.show()
```

```python
plt.figure(figsize=(10,7))
corr = df.select_dtypes(include="number").corr()
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

```python
cols = ["Customer_care_calls","Cost_of_the_Product","Discount_offered","Weight_in_gms","Reached.on.Time_Y.N"]
sns.pairplot(df[cols], hue="Reached.on.Time_Y.N")
plt.show()
```

```python
print(df.groupby("Reached.on.Time_Y.N")[["Discount_offered","Weight_in_gms","Cost_of_the_Product"]].mean())
```

```python
print(df.groupby(["Mode_of_Shipment","Reached.on.Time_Y.N"]).size().unstack())
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Reached.on.Time_Y.N", y="Discount_offered", data=df)
plt.title("Average Discount Offered vs On-Time Delivery")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x="Mode_of_Shipment", y="Weight_in_gms", data=df)
plt.title("Average Weight by Mode_of_Shipment")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

df = pd.read_csv("Train.csv")

```

```python
ct = pd.crosstab(df["Mode_of_Shipment"],
                 df["Reached.on.Time_Y.N"],
                 normalize="index")

ct.plot(kind="bar", stacked=True, figsize=(8,5))
plt.title("Mode_of_Shipment vs On-Time Delivery (Percentage)")
plt.ylabel("Percentage")
plt.show()
```

```python
summary_box = df.groupby("Reached.on.Time_Y.N").agg({
    "Discount_offered":"mean",
    "Weight_in_gms":"mean",
    "Customer_care_calls":"mean",
    "Prior_purchases":"mean"
}).round(2)

summary_box
```

```python
summary_box.style.set_caption("Summary Statistics by Delivery Status")\
                 .background_gradient(cmap="Blues")\
                 .set_properties(**{'border': '2px solid black'})
```

```python
ct_box = pd.crosstab(df["Mode_of_Shipment"],
                     df["Reached.on.Time_Y.N"],
                     margins=True)

ct_box
```

```python
stats = df["Discount_offered"].describe().round(2)

plt.figure(figsize=(6,4))
plt.text(0.1, 0.5, stats.to_string(),
         bbox=dict(facecolor='lightblue', edgecolor='black'))
plt.axis('off')
plt.title("Discount Offered Statistics")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.boxplot(x="Reached.on.Time_Y.N",
            y="Discount_offered",
            data=df)

plt.title("Discount Offered vs Delivery Status")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.xticks(rotation=90)
plt.title("Boxplot of All Numerical Features")
plt.show()
```
